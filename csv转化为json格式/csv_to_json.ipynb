```python
import csv
import json
```

```python
include_columns = ["user", "platform", "start_lon", "start_lat", "end_lon", "end_lat"]

rows = []
with open("bike_od.csv") as f:
    for row in csv.DictReader(f):
        row = {c: row[c] for c in row if c in include_columns}
        rows.append(row)

with open("bike_od2.js", "w") as js_file:
    js_file.write("let bike_ods = ")
    json.dump(rows, js_file)
```
